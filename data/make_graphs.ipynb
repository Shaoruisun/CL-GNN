```python
import oddt
import oddt.pandas as opd
from oddt.fingerprints import PLEC
import numpy as np
from oddt.fingerprints import SPLIF
import numpy as np
import pandas as pd
from mol2graph import pdbbind_handle
import os

import multiprocessing

filelist = os.listdir('/data/new_data/data_2')
protein_path='data/new_data/data_2'
ligand_path='/data/new_data/data_2'
results = []
fingerprints =[]
error=[]
#np.load(ids)
```

```python
ids_data=[]
for index,i in enumerate(filelist):
    #print(index)
    if os.path.exists("%s/%s/out_id.npy"%(protein_path,i))==True:
        ids=np.load("%s/%s/out_id.npy"%(protein_path,i))
        #graphs_p=np.load("%s/%s/out_protein.npy"%(protein_path,i)
        #graphs_l=np.load("%s/%s/out_ligand.npy"%(protein_path,i))
        #fingerprints=np.load("%s/%s/out_PLEC.npy"%(protein_path,i))
        ids_data.append(ids[0])
    else:
        print(index)
np.save("out_id_6.21.npy", ids_data)
```

```python
from dgl.data.utils import save_graphs
graphs_p_data=[]
for index,i in enumerate(filelist):
    #print(index)
    if os.path.exists("%s/%s/out_id.npy"%(protein_path,i))==True:
        graphs_p=np.load("%s/%s/out_protein.npy"%(protein_path,i))
        #fingerprints=np.load("%s/%s/out_PLEC.npy"%(protein_path,i))
        graphs_p_data.append(graphs_p)
    else:
        print(index)
save_graphs("out_protein_6.21.bin", graphs_p_data)
```

```python
graphs_l_data=[]
for index,i in enumerate(filelist):
    #print(index)
    if os.path.exists("%s/%s/out_id.npy"%(protein_path,i))==True:
        #ids=np.load("%s/%s/out_id.npy"%(protein_path,i))
        graphs_l=np.load("%s/%s/out_ligand.npy"%(protein_path,i))
        graphs_l_data.append(graphs_l)
    else:
        print(index)
save_graphs("out_ligand_6.21.bin", graphs_l_data)
```
